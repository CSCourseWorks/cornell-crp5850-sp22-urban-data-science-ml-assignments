<h1> Exercise 1 </h1>
<p> Create a figure object called fig using plt.figure()

Use add_axes to add an axis to the figure canvas at [0,0,1,1]

Plot (x,y) on that axes and set the labels and titles to match the plot below:</p>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 100, 100)
y = x**2
z = x**3
```

```python
fig1 = plt.figure()
axes = fig1.add_axes([0,0,1,1])
axes.plot(x, y)
axes.set_xlabel('x')
axes.set_ylabel('x square')
axes.set_title('Exercise 1')
```

<h1> Exercise 2 </h1>
<p> Create a figure object and put two axes on it, located at [0,0,1,1] and [0.1,0.4,0.5,0.5] respectively.

Plot x,y on the first axes, and plot x,z on the second axes, set labels, titles, and color to match the plot below </p>

```python
fig2 = plt.figure()
ax1 = fig2.add_axes([0,0,1,1])
ax2 = fig2.add_axes([0.1,0.4,0.5,0.5])
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('x square')
ax1.set_title('Figure 1')
ax2.plot(x, z, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('x cubed')
ax2.set_title('Figure 2')
```

<h1> Exercise 4 </h1>
<p> Use plt.subplot to create the figure below

figure size is 12, 4

the first figure: plot (x,y), color is blue, linewidth is 5, line type is dashline ('--'). legend is 'x squared'

the second figure: plot(x,z), color is red, linewidth is 2. legend is 'x cubed' </p>

<h3> Answer 1 </h3>
<p> Using plt.figure and plt.subplot </p>

```python
fig4 = plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(x, y, color='b', linewidth=5, linestyle='--', label='x squared')
plt.legend(loc='upper left')
plt.subplot(1, 2, 2)
plt.plot(x, z, color='r', linewidth=2, label='x cubed')
plt.legend(loc='upper left')
```

<h3> Answer 2 </h3>
<p> Using plt.subplots </p>

```python
fig4, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(x, y, color='b', linewidth=5, linestyle='--', label='x squared')
axes[0].legend(loc='upper left')
axes[1].plot(x, z, color='r', linewidth=2, label='x cubed')
axes[1].legend(loc='upper left')
```

<h1> Seaborn Exercise </h1>

```python
import matplotlib.pyplot as plt
import seaborn as sns
mpg = sns.load_dataset('mpg')
```

<h1> Exercise 1 </h1>
<p> Draw a histogram on mpg, bin=20 </p>

```python
sns.histplot(data=mpg['mpg'], bins=20)
```

<h1> Exercise 2 </h1>
<p> Compare the distribution between mpg and horsepower </p>

```python
sns.jointplot(x='mpg', y='horsepower', data=mpg)
```

<h1> Exercise 3 </h1>
<p> Compare the distribution between mpg and horsepower by origin </p>

```python
sns.jointplot(x='mpg', y='horsepower', data=mpg, hue='origin')
```

<h1> Exercise 4 </h1>
<p> Draw a heatmap to show the correlation between variables the color is 'coolwarm'
the titel is 'correlation analysis' </p>

```python
g = sns.heatmap(mpg.corr(), cmap='coolwarm', annot=True)
g.set_title('correlation analysis')
```

<h1> Exercise 5 </h1>
<p> Use FacetGrid to show the relation between mpg and horsepower by the types of cylinders and by the origins </p>

```python
g = sns.FacetGrid(data=mpg, col='cylinders', row='origin')
g.map(plt.scatter, 'mpg', 'horsepower')
```
